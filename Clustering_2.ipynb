```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
data = pd.read_csv('DataSet.csv')

xnya = (data['x'])
ynya = (data['y'])
clas = (data['kelas'])

xn = tuple(xnya)
yn = tuple(ynya)
celas = tuple(clas)

datset = np.array([xn,yn]),np.array(celas)
```

```python
points =  datset[0]
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dendrogram = sch.dendrogram(sch.linkage(points, method = 'ward'))
```

```python
plt.scatter(datset[0][:,0], datset[0][:,1])
```

```python
hc = AgglomerativeClustering(n_clusters =2, affinity = 'euclidean', linkage = 'ward')
```

```python
y_hc = hc.fit_predict(points)
```

```python
plt.scatter(points[y_hc == 0,0], points[y_hc == 0,1], s = 300, c = 'cyan')
plt.scatter(points[y_hc == 1,0], points[y_hc == 1,1], s = 300, c = 'red')

plt.scatter
```
